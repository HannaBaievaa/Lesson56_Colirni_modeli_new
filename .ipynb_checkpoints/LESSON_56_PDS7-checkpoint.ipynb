```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im = cv2.imread("lorikeet.jpg")
```

```python
type(im)
```

```python
im
```

```python
plt.imshow(im)
plt.show
```

```python
im_RGB = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im_RGB)
plt.show
```

```python
im_RGB
```

```python
R = im_RGB.copy()
```

```python
R[:,:,(1,2)]=0
plt.imshow(R)
plt.show
```

```python
G = im_RGB.copy()
G[:,:,(0,2)]=0
plt.imshow(G)
plt.show
```

```python
B = im_RGB.copy()
B[:,:,(0,1)]=0
plt.imshow(B)
plt.show
```

```python
img_HSV = cv2.cvtColor(im_RGB, cv2.COLOR_RGB2HSV)
plt.imshow(img_HSV)
plt.show
```

```python
img_GREY = cv2.cvtColor(im_RGB, cv2.COLOR_RGB2GRAY)
plt.imshow(img_GREY)
plt.show
```
