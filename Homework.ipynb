```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("cat.jpg")
```

```python
plt.imshow(image)
plt.show
```

### Декомпозиція по складових RGB-моделі

```python
# конвертуємо картинку в RGB
im_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(im_RGB)
plt.show
```

```python
#виділяємо червоний колір
R = im_RGB.copy()
R[:,:,(1,2)]=0
plt.imshow(R)
plt.show
```

```python
# виділяємо зелений колір
G = im_RGB.copy()
G[:,:,(0,2)]=0
plt.imshow(G)
plt.show
```

```python
# виділяємо синій(блакитний) колір
B = im_RGB.copy()
B[:,:,(0,1)]=0
plt.imshow(B)
plt.show
```

### Декомпозиція по складових HSV-моделі

```python
# конвертуємо картинку в HSV
img_HSV = cv2.cvtColor(im_RGB, cv2.COLOR_RGB2HSV)
plt.imshow(img_HSV)
plt.show
```

```python
H = img_HSV.copy()
H[:,:,(1,2)]=255
H_RGB = cv2.cvtColor(H, cv2.COLOR_HSV2RGB)
plt.imshow(H_RGB)
plt.show()
```

```python
S = img_HSV.copy()
S[:,:,(0,2)]=255
S_RGB = cv2.cvtColor(S, cv2.COLOR_HSV2RGB)
plt.imshow(S_RGB)
plt.show()
```

```python
V = img_HSV.copy()
V[:,:,(0,1)]=255
V_RGB = cv2.cvtColor(V, cv2.COLOR_HSV2RGB)
plt.imshow(V_RGB)
plt.show()
```
